```python
df_music_segment = df["Music effects"].copy()
df_music_segment
```

```python
df["Music effects"]
```

```python

for i in range(len(df_music_segment)):
    if df_music_segment[i] == "Worsen":
        df_music_segment[i]  = 1.
    elif df_music_segment[i] == "Improve":
        df_music_segment[i]  = 2.
    elif df_music_segment[i] == "No effect":
        df_music_segment[i]  = 3.

df_music_segment
df["df_music_segment"] = df_music_segment
df.head()
```

```python
df_2 = df[["Age","Hours per day","BPM","Anxiety","Depression","Insomnia","OCD","df_music_segment"]]
```

```python
correlation_2 = df_2.corr()

custom_palette = sns.light_palette("#3B687B", as_cmap=True)

sns.heatmap(correlation_2.abs(), annot=True, cmap = custom_palette)
plt.show()
```

```python
anxiety_high = df.loc[(df['Anxiety'] > 8)].mean()[1]
depression_high = df.loc[(df['Depression'] > 8)].mean()[1]
insomnia_high = df.loc[(df['Insomnia'] > 8)].mean()[1]
ocd_high = df.loc[(df['OCD'] > 8)].mean()[1]

high_means = [anxiety_high, depression_high, insomnia_high, ocd_high]

m_all = ["Anxiety", "Depression", "Insomnia", "OCD"]
custom_palette = sns.light_palette("#3B687B", as_cmap=True)

plt.barh(m_all, high_means, color= "#3B687B")
  
plt.xlabel("Durée moyenne d'écoute en H")
plt.title("Durée moyenne d'écoute pour les taux de MH elevé")
plt.xlim(3,5.5)
plt.show()
```

```python
anxiety_low = df.loc[(df['Anxiety'] < 3)].mean()[1]
depression_low = df.loc[(df['Depression'] < 3)].mean()[1]
insomnia_low = df.loc[(df['Insomnia'] < 3)].mean()[1]
ocd_low = df.loc[(df['OCD'] < 3)].mean()[1]

low_means = [anxiety_low, depression_low, insomnia_low, ocd_low]

plt.barh(m_all, low_means, color="#3B687B")
  
plt.xlabel("Durée moyenne d'écoute en H")
plt.title("Durée moyenne d'écoute pour les taux de MH bas")
plt.xlim(3,4)
plt.show()
```
