```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

y =df["Music effects"]
#X = df['Fav genre']
X = pd.get_dummies(df['Fav genre'], prefix='Genre')
X["Anxiety"] = df["Anxiety"]
X["Depression"] = df["Depression"]
X["Insomnia"] = df["Insomnia"]
X["OCD"] = df["OCD"]
#X = df.drop("Music effects",axis=1)

X_train , X_test, y_train, y_test = train_test_split(X,y, test_size = 0.15, random_state=2023)
```

```python
clf = RandomForestClassifier()
clf.fit(X_train,y_train)
predict = clf.predict(X_test)
accuracy = accuracy_score(y_test,predict)
print("Accuracy: ",accuracy)
print("Rapport de classification : ")
print(classification_report(y_test,predict))
```
